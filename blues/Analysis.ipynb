```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#'Iter,Accept,Theta,Work,LogP,Dist,State'
```

```python
v1 = pd.read_csv('bluesv1-data5.txt', skipfooter=1)
v2 = pd.read_csv('bluesv2-data5.txt', skipfooter=1)
```

```python
def labelBars(ax):
    # create a list to collect the plt.patches data
    totals = []

    # find the values and append to list
    for i in ax.patches:
        totals.append(i.get_height())

    # set individual bar lables using above list
    total = sum(totals)

    # set individual bar lables using above list
    for i in ax.patches:
        # get_x pulls left or right; get_height pushes up or down
        ax.text(i.get_x()+.20, i.get_height(), \
                str(round((i.get_height()/total)*100, 2))+'%', fontsize=15)
    return ax
```

```python
fig, ax = plt.subplots(2, figsize=(16,9))

freq1 = v1['State'].value_counts()
freq2 = v2['State'].value_counts()

freq1.plot.bar(ax=ax[0], label='State v1')
freq2.plot.bar(ax=ax[1], label='State v2')

ax[0] = labelBars(ax[0])
ax[1] = labelBars(ax[1])

plt.autoscale()
```

```python
fig, ax = plt.subplots(2, figsize=(16,9), sharex=True)
sns.distplot(v1['Work'], ax=ax[0], axlabel='Work v1')
sns.distplot(v2['Work'], ax=ax[1], axlabel='Work v2')
#ax[1].set_xlim((-0.02, 0.02))
fig.tight_layout()
```
